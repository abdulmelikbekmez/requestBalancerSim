```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cikti.csv")
c1 = df[df["client id"] == 1]
c2 = df[df["client id"] == 2]
```

```python
plt.plot(c1["time"],c1["client id"],"bo", c2["time"], c2["client id"], "ro")
plt.xlabel("time")
plt.ylabel("client id")
#2 clients, one has higher prio and long job other has lower prio and short job
```

```python
plt.plot(c1["duration"],c1["total waiting time"],"ro", c2["duration"], c2["total waiting time"], "bo")
plt.xlabel("duration")
plt.ylabel("request's waiting time")
#2 clients one has higher prio and long job other has lower prio and short job
```

```python
df = pd.read_csv("cikti.csv")
c1 = df[df["client id"] == 1]
c2 = df[df["client id"] == 2]
plt.plot(c1["time"],c1["client id"],"bo", c2["time"], c2["client id"], "ro")
plt.xlabel("time")
plt.ylabel("client id")
plt.title("Total Request vs Completed Jobs")
```
